This URL is for software job ads in Boston published on Craigslist:

https://boston.craigslist.org/search/sof

Using BeautifulSoup and Requests, scrape the title and full URL of each job ad. The result should look like this:

Job: --------------

URL: --------------


Job: --------------

URL: --------------



Job: --------------

URL: --------------

```python
from bs4 import BeautifulSoup
import requests
import re

url = 'https://boston.craigslist.org/search/sof'

response = requests.get(url)
 
# Extracting the source code of the page.
data = response.text
 
# Passing the source code to Beautiful Soup to create a BeautifulSoup object for it.
soup = BeautifulSoup(data, 'html.parser')
#print(soup.prettify())

job_list = soup.find_all('a',class_='result-title hdrlnk')

for job in job_list:
    print('Job:',job.string)
    print('URL:',job['href'])
    print('\n')
```
