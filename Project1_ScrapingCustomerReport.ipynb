## Getting list of product

```python
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import requests
import re

# Background on user-agents
url = 'https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
ua = UserAgent()

header = {'user-agent':ua.chrome}

page = requests.get(url,headers=header)

soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())

product_tags = soup.find_all('a',{"class":"products-a-z__results__item"})

# write product name to file
file_name = 'consumer_reports.txt'
raw_html_file = 'consumer_html.txt'

with open(raw_html_file, 'w') as file:
    file.write(page.content.decode('utf-8') if type(page.content) == bytes else file.write(page.content))

with open(file_name, 'w') as file:
    for product in product_tags:
        print(product['href'])
        file.write(product.string + ':' + product['href'] )
        
```
